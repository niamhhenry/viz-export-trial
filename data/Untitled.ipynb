```python
import pandas as pd
import plotly as py
import plotly.graph_objs as go
import plotly.express as px

from PIL import Image
```

```python
df=pd.read_csv('pax.csv')
```

```python
df['Dat']=pd.to_datetime(df['Dat'])
```

```python
df['year']=pd.DatetimeIndex(df['Dat']).year
```

```python
grouped_yr=df.groupby('year')
```

```python
nr_agts_yr=pd.DataFrame(grouped_yr['AgtId'].nunique()).rename(columns={'AgtId':'Nr Agts'}).reset_index()
```

```python
fig = px.bar(nr_agts_yr, x='year', y="Nr Agts")
Logo = Image.open("PeaceRep-Logo-480x151.png")

fig.add_layout_image(
    dict(
        source=Logo,
        xref="paper", yref="paper",
        x=1, y=1.05,
        sizex=0.2, sizey=0.2,
        xanchor="right", yanchor="bottom"
    )
)
fig.update_layout(
    hovermode="x",
    title=("PA-X Peace Agreements Database<br>" +
           "<i>Number of Agreements per Year</i>"),
)
fig.show()
```

```python
fig.write_image("number_agts_year.png")
```
